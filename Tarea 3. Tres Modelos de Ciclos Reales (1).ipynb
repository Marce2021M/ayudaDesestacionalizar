## Un modelo con producción doméstica
Considere una economía compuesta por un hogar representativo dotado de preferencias sobre consumo de mercado $c_{mt}$, servicios de consumo doméstico $c_{dt}$. Además el hogar debe decidir la fracción de sus miembros que trabajará en el mercado $\psi_{t}$ y la fracción que trabajará en el hogar $\eta_{t}$. Ambos tipos de trabajo generan el mismo nivel de desutilidad $B$. Podemos escribir matemáticamente lo descrito como
$$U=E_{0}\left\{\sum^{\infty}_{t=0}\beta^{t}\left[\log(c_{mt})+\mu\log(c_{dt})-B\left(\psi_{t}+\eta_{t}\right)\right]\right\}$$
En esta economía hay dos tecnologías. Una de producción de bienes de mercado
$$y_{t}=A_{t}k^{\alpha}_{t}l^{1-\alpha}_{mt}$$
donde 
$$k_{t+1}=(1-\delta_{m})k_{t}+i_{mt}$$
y una función de producción de servicios domésticos 
$$c_{dt}=\left[\nu k^{\sigma}_{dt}+(1-\nu)(\eta_{t}\bar{h})^{\sigma} \right]^{\frac{1}{\sigma}}$$ donde 
$$k_{dt+1}=(1-\delta_{d})k_{dt}+q_{t}i_{dt}$$
Tanto la productividad total de los factores $A_{t}$, como la calidad de la inversión en bienes duraderos $q_{t}$ están sujetas a perturbaciones aleatorias que podemos escribir como 
$$\log(A_{t+1})=\rho_{A}\log(A_{t})+\epsilon_{At+1}$$
donde $\epsilon_{At+1}\longrightarrow N(0,\sigma^{2}_{A})$ y
$$\log(q_{t+1})=\rho_{q}\log(q_{t})+\epsilon_{qt+1}$$
donde $\epsilon_{qt+1}\longrightarrow N(0,\sigma^{2}_{q})$.

Dado que no hay nada que impida el cumplimiento de los teoremas fundamentales de la economía del bienestar, el problema se trata de resolver el siguiente problema de planificación social

$$\max_{c_{mt},c_{dt},\psi_{t},\eta_{t},k_{t+1},k_{dt+1}}E_{0}\left\{\sum^{\infty}_{t=0}\beta^{t}\left[\log(c_{mt})+\mu\log(c_{dt})-B\left(\psi_{t}+\eta_{t}\right)\right]\right\}$$
sujeto a:
$$c_{mt}+k_{t+1}-(1-\delta)k_{t}+\frac{k_{dt+1}-(1-\delta_{d})k_{dt}}{q_{t}}=A_{t}k^{\alpha}_{t}(\psi_{t}\bar{h})^{1-\alpha}$$ a la tecnología de producción doméstica
$$c_{dt}=\left[\nu k^{\sigma}_{dt}+(1-\nu)(\eta_{t}\bar{h})^{\sigma}_{t} \right]^{\frac{1}{\sigma}}$$ y a la evolución estocástica de la productividad total de los factores y la calidad de los bienes duraderos descrita anteriormente.

Vamos a suponer, salvo que se indique lo contrario, que $\mu=.3$, $B=1.5$, $\beta=.98$, $\bar{h}=.33$, $\alpha=.5$, $\nu=.2$, $\sigma=-.5$, $\rho_{A}=.95$, $\rho_{q}=.6$, $\sigma^{2}_{A}=.007$, $\sigma^{2}_{q}=.01$. Además estamos suponiendo que los valores de estos choques en un hipotético estado estacionario están normalizados a $q=A=1$.

## Un modelo con choques energéticos a empresas y hogares
Considere una economía poblada por un hogar representativo que valora el consumo $c_{t}$ pero obtiene una desutilidad $B$ que experimentan las personas del hogar que tienen que it a trabajar en el mercado laboral. Escribimos estas preferencias como 
$$U=E_{0}\left\{\sum^{\infty}_{t=0}\beta^{t}\left[\log(c_{t})-B\psi_{t}\right]\right\}$$
Esta economía produce un bien $y_{t}$ mediante el uso de capital $k_{t}$, trabajo $l_{t}$ y energía $e_{t}$, que se combinan mediante la tecnología
$$y_{t}=A_{t}\left[\alpha k^{\sigma}_{t}+\eta l^{\sigma}_{t}+(1-\alpha-\eta)e^{\sigma}_{t}\right]^{\frac{1}{\sigma}}$$
donde $\alpha$, $\eta$ son pesos de los respectivos insumos en el producción y $\frac{1}{1-\sigma}$ es la elasticidad de sustitución entre los insumos y lo lógico es que sean complementos. El capital se acumula mediante un proceso geométrico estandard
$$k_{t+1}=(1-\delta)k_{t}+i_{t}$$
El precio de la energía $p_{et}$ sigue un proceso estocástico que hogares y empresas toman como dado y sobre el que no pueden influir
$$\log(p_{et+1})=(1-\rho_{e})\log(p_{e})+\rho_{e}\log(p_{et})+\epsilon_{et+1}$$
y la productividad con la que se combinan los tres insumos también está sujeta a choques aleatorios tal que 
$$\log(A_{t+1})=\rho_{A}\log(A_{t})+\epsilon_{At+1}$$

Dado que se cumplen los teoremas fundamentales de la economía del bienestar podemos escribir el problema de planificación social como: 

$$\max_{c_{t},\psi_{t},e_{t},k_{t+1}}E_{0}\left\{\sum^{\infty}_{t=0}\beta^{t}\left[\log(c_{t})-B\psi_{t}\right]\right\}$$
sujeto a:
$$c_{t}+k_{t+1}-(1-\delta)k_{t}+p_{et}(e_{t}+\psi_{t}\bar{e})=A_{t}\left[\alpha k^{\sigma}_{t}+\eta \psi^{\sigma}_{t}+(1-\alpha-\eta)e^{\sigma}_{t}\right]^{\frac{1}{\sigma}}$$
y a los procesos estocásticos especificados anteriormente.

Suponga, salvo que se le indique lo contrario que los parámetros del modelo son $B=1.5$, $\beta=.98$, $\alpha=.3$, $\eta=.5$, $\sigma=-.5$, $\delta=.025$, $p_{e}=1$, $\bar{e}=.01$ $\rho_{A}=.96$, $\rho_{e}=.7$, $\sigma_{A}=0.007$ y $\sigma_{e}=0.015$.  

### Preguntas específicas del modelo
1. Represente la cantidad de trabajo $l$ y la ratio de capital producto $\frac{k}{y}$ en el estado estacionario cuando varía el peso que tiene la energía en la tecnología de producción $1-\alpha-\eta$ entre $.1$ y $.4$. 

2. Represente la función de impulso respuesta de la cantidad de trabajo $\psi_{t}$ y el gasto real en energía $e_{t}$ ante un incremento transitorio en t=0 del 15% al precio de la energía en un gráfico de 1x2

3. Represente la función de impulso respuesta de la cantidad de trabajo $\psi_{t}$ y el gasto real en energía $e_{t}$ ante una reducción transitoria en t=0 del 5% a la productividad total de los factores en un gráfico de 1x2

4. Represente la función de impulso respuesta de la cantidad de trabajo y del consumo real de energía para los valores de $\bar{e}\in\{0, .1, .2\}$ de tal forma que los valores de $\beta$ y $B$ se ajusten para que la ratio de capital producto y la cantidad de trabajo sean los mismos que la que surge de los parámetros dados en el enunciado. Discuta que papel juega el que los hogares afronten un coste fijo por ir al trabajo

5. Obtenga la desviación típica relativa y la correlación respecto del PIB de la economía de: el empleo $\psi_{t}$, el gasto real en energía $p_{et}(e_{t}+\psi_{t}\bar{e})$, el consumo $c_{t}$ y la inversión $i_{t}$ para los valores de los parámetros del enunciado de esta pregunta 

#### Posibles extensiones de este modelo
Este modelo podría extenderse para pensar en cuáles serían las ganancias de bienestar de políticas para estabilizar el gasto energético de los hogares y, en ese contexto, tiene sentido pensar en el modelo con algún elemento de producción doméstica.

Otra alternativa es pensar este modelo en un contexto multisectorial y con elementos que permitan que la política monetaria sea relevante para poder responder a preguntas relativas a el comportamiento de la inflación causada por choques energéticos y las consecuencias para el bienestar de intentar frenarla con una política monetaria restrictiva.

## Un modelo con informalidad y política fiscal
Considere una economía poblada por un hogar representativo que está dotado con preferencias sobre el consumo privado $c_{t}$, el consumo de un bien público $G_{t}$ y que obtiene la misma desutilidad constante $B$ de mandar a parte de los hogares a trabajar en el sector formal $\psi_{t}$ o al sector informal $\eta_{t}$. La parte del hogar en el sector formal tiene que pagar impuestos sobre su renta laboral, pero tiene acceso a un bien público $G_{t}$. Los miembros del hogar en el sector informal o que no parcipan en el mercado laboral no pagan impuestos pero tampoco pueden disfrutar del bien público. El hogar en su conjunto paga el impuesto a la renta del capital que no es posible evadir ocultándose en el sector informal.

El gobierno presupuesta el bien público como una tasa $g$ sobre el valor del pib de la economía pero puede haber desviaciónes causadas por errores o retrasos en la ejecución del objetivo de gasto anunciado, mecanismo que podemos describir mediante la siguiente ecuación
$$\log(g_{t+1}) = (1-\rho_{g})\log(g)+\rho_{g}\log(g_{t})+\epsilon_{gt+1}$$ donde $\epsilon_{gt+1}\longrightarrow N(0,\sigma^{2}_{g})$. 

Para poder gastar $g$ en promedio el gobierno cobra un impuesto sobre la renta promedio de $\tau$, pero también puede haber desviaciones que quedan descritas en el siguiente proceso estocástico
$$\log(\tau_{t+1}) = (1-\rho_{\tau})\log(\tau)+\rho_{\tau}\log(\tau_{t})+\epsilon_{\tau t+1}$$
donde $\epsilon_{\tau t+1}\longrightarrow N(0,\sigma^{2}_{\tau})$. 

Lo que sobra de estos procesos de gastos y recaudación es devuelto a los hogares como una transferencia de suma fija y el gobierno no incurre en ningún tipo de endeudamiento, por lo que su restricción presupuestaria puede escribirse como:
$$\Omega^{*}_{t}+g_{t}y_{t}=\tau_{t}(w_{t}\psi_{t}+q_{t}k_{t})$$

Por último, una empresa representativa produce el único bien de la economía $y_{t}$, mediante una tecnología Cobb-Douglas 
$$y_{t}=A_{t}k^{\alpha}_{t}l^{1-\alpha}_{t}$$
donde $k_{t+1}=(1-\delta)k_{t}+i_{t}$ y $A_{t}$ está sujeto tambien a choques aleatorios que quedan descritos en el siguiente proceso 
$$\log(A_{t+1}) = \rho_{A}\log(A_{t})+\epsilon_{At+1}$$
donde $\epsilon_{At+1}\longrightarrow N(0,\sigma^{2}_{A})$.

Asumimos que los tres procesos estocásticos descritos son independientes entre sí. Todas estas palabras y lo que no haya mencionado explícitamente se resumen en el problema de decisión del hogar y de la empresa representativa:
$$\max_{c_{t},\psi_{t},\eta_{t},k_{t+1}}E_{0}\left\{\sum^{\infty}_{t=0}\beta^{t}\left[\psi_{t}\log(c_{t}+\mu G_{t})+(1-\psi_{t})\log(c_{t})-B(\psi_{t}+\eta_{t})\right]\right\}$$
sujeto a:
$$c_{t}+k_{t+1}-(1-\delta)k_{t}=(1-\tau_{t})(w_{t}\psi_{t}+q_{t}k_{t})+w_{t}\eta_{t}+\Omega^{*}_{t}$$

Suponga, salvo que se le indique lo contrario, los siguientes parámetros del modelo $\mu=.8$, $\beta=.98$, $g=.10$, $\alpha=.5$, $\delta=.025$, $\rho_{A}=.8$, $\rho_{g}=.95$, $\rho_{\tau}=.95$, $\sigma_{A}=.007$, $\sigma_{g}=.001$, $\sigma_{\tau}=.001$. 

### Preguntas específicas de este modelo
1. Represente la cantidad de personas empleadas $\psi+\eta$, la tasa de informalidad $\frac{\eta}{\psi+\eta}$ respecto del empleo total y la tasa de informalidad respecto de la población $\eta$ en el estado estacionario cuando $g$ varía entre $.01$ y $.25$ en un gráfico de 1x3

4. #### Posibles extensiones de este modelo
Una descripción más realista de la economía debería tener en cuenta el lado de la búsqueda y de la creación de empleo con protección social o no, así que piensen en la tarea como una prueba de concepto y no como una descripción seria de un fenómeno como el comportamiento cíclico de la informalidad. Las empresas podrían jugar algún papel en la determinación del tamaño del sector informal. Este modelo podría utilizarse para estudiar cuánto de la política fiscal puede explicar las fluctuaciones en el sector informal.
